```python
scl = [
    [0.0, 'rgb(142,140,247)'],
    [0.2, 'rgb(218,218,235)'],
    [0.4, 'rgb(188,189,220)'],
    [0.6, 'rgb(158,154,200)'],
    [0.8, 'rgb(117,107,177)'],
    [1.0, 'rgb(84,39,143)']
]
```

```python
data = [go.Choropleth(
    colorscale = scl,
    autocolorscale = False,
    locations = df_plot['code'],
    z = df_plot['BIRTHS2010'].astype(float),
    locationmode = 'USA-states',
    text = df_plot['text'],
    marker = go.choropleth.Marker(
        line = go.choropleth.marker.Line(
            color = 'rgb(255,255,255)',
            width = 2
        )),
    colorbar = go.choropleth.ColorBar(
        title = "Thousands USD")
)]
```

```python
layout = go.Layout(
    title = go.layout.Title(
        text = 'Diabetes Prevalance in the US by State<br>(Hover for breakdown)'
    ),
    geo = go.layout.Geo(
        scope = 'usa',
        projection = dict(type = 'albers usa'),
        showlakes = True,
        lakecolor = 'rgb(66, 165, 245)'),
)
```
