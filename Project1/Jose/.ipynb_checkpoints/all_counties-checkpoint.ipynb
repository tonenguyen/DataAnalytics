## First solve for 2 states

```python
for index,row in states_df.iterrows():
    state = row['State']
    county = row['County']
    abrev = abrev_df.loc[state][0]
    states_df.loc[index, 'code'] = abrev
    print(f'{state}: {county} abreviation: {abrev}')
    #states2010.loc[index, 'FIPS'] = code
    #print(f'{state}: {county} abreviation: {abrev}')
```

```python
scl = [
    [0.0, 'rgb(142,240,247)'],
    [0.2, 'rgb(218,218,235)'],
    [0.4, 'rgb(188,189,220)'],
    [0.6, 'rgb(158,154,200)'],
    [0.8, 'rgb(117,107,177)'],
    [1.0, 'rgb(84,39,143)']
]
```

```python
states_df = states_df[~(states_df['percent'] == 'No Data')]
```

```python
data = [go.Choropleth(
    colorscale = 'Blues',
    autocolorscale = False,
    locations = states_df['code'],
    z = states_df['percent'].astype(float),
    locationmode = 'USA-states',
    text = states_df['State'],
    marker = go.choropleth.Marker(
        line = go.choropleth.marker.Line(
            color = 'rgb(255,255,255)',
            width = 2
        )),
    colorbar = go.choropleth.ColorBar(
        title = "Thousands USD")
)]
```

```python
layout = go.Layout(
    title = go.layout.Title(
        text = 'Diabetes Prevalance in the US by State<br>(Hover for breakdown)'
    ),
    geo = go.layout.Geo(
        scope = 'usa',
        projection = dict(type = 'albers usa'),
        showlakes = True,
        lakecolor = 'rgb(66, 165, 245)'),
)
```

```python
#fig = dict(data=data, layout=layout)
fig = go.Figure(data = data, layout = layout)
py.iplot(fig, filename='d3-cloropleth-map')
```
